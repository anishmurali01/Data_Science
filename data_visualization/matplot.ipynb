```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,3,3]
y=[3,4,5,2]
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
X=np.array([1,2,3,4,5,6,7,8])
y1 = 2*X+3
y2 = X**2 - 3*X + 5
plt.plot(X,y1)
plt.plot(X,y2)

plt.title('Work vs Time')
plt.xlabel('Time')
plt.ylabel('Work')
plt.ylim(0,30)
plt.xlim(0,6)
plt.grid()
plt.show()
```

```python
x = np.arange(1,11)
y1 = [10,5,15,20,13,16,27,18,23,30]
y2 = 2*x+3
plt.plot(x,y1,'ro-',label = 'Price')
plt.plot(x,y2,'gx-.',label = 'Line')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
x = np.arange(1,11)
y1 = 3*x+5
y2 = 2*x+3
plt.plot(x,y1,label = 'Price',color = 'r',marker = '*',markersize = 15,markerfacecolor = 'b')
plt.plot(x,y2,label = 'Line',color = 'g',marker = 'o',linestyle=':')

plt.xlabel('X',fontdict=font)
plt.ylabel('f(X) = Y',fontdict = font)
plt.title("Line Plot",fontdict = font)
plt.legend()
plt.grid()
#plt.savefig('LinePlot.jpg')
plt.show()
```

```python
X = np.linspace(-3,3,100)
y1 = 3*X+5
y2 = X**2+5

lines = plt.plot(X,y1,X,y2)

plt.setp(lines[0],color = 'r')
plt.setp(lines[1],color = 'g')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
# plt.xlim((-4,4))
# plt.ylim((2.5,12))
plt.grid()
plt.show()
print(lines)
```

```python
X = np.linspace(-3,3,10)
y1 = 3*X+5
y2 = X**2+5

plt.scatter(X,y1,marker = '*',s = 30,c = 'r',label='Linear')
plt.scatter(X,y2,marker = '^',s = 40,c = 'blue',label='Quadratic')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
plt.legend()
plt.grid()
plt.show()
```

```python
X = np.linspace(-2,2,1000)
y = np.sin(2*np.pi*X)

plt.figure(figsize=(10,2))
plt.plot(X,y)
plt.grid()
plt.title("Sin Wave")
plt.show()
```

```python
x= np.linspace(0, 2 * np.pi, 1000)
y = np.cos(x)

plt.plot(x, y, label='cos(x)')
plt.title('Cos Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.tan(x)

y[np.abs(y) > 10] = np.nan

plt.plot(x, y, label='tan(x)', color='orange')
plt.title('Tangent Wave')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.grid(True)
plt.legend()
plt.ylim(-10, 10)
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = 1 / np.tan(x)

y[np.abs(y) > 10] = np.nan

plt.plot(x, y, label='cot(x)', color='green')
plt.title('Cotangent Wave')
plt.xlabel('x')
plt.ylabel('cot(x)')
plt.grid(True)
plt.legend()
plt.ylim(-10, 10)
plt.show()
```
